```python
import pandas as pd
import csv
```

```python
path = 'output_csv_file.csv'

df_original = pd.read_csv(path, sep=',', quoting=csv.QUOTE_ALL, escapechar='\\')
```

```python
df = df_original.copy()

LIWC_keys = ['text_WC','text_Funct','text_TotPron','text_PronPer','text_Yo','text_Nosotro','text_TuUtd','text_ElElla','text_Ellos','text_PronImp','text_Articulo','text_Verbos','text_VerbAux','text_Pasado','text_Present','text_Futuro','text_Adverb','text_Prepos','text_Conjunc','text_Negacio','text_Cuantif','text_Numeros','text_Maldec','text_verbYO','text_verbTU','text_verbNOS','text_verbosEL','text_verbELLOS','text_Subjuntiv','text_VosUtds','text_formal','text_informal','text_verbVos','text_Social','text_Familia','text_Amigos','text_Humanos','text_Afect','text_EmoPos','text_EmoNeg','text_Ansiedad','text_Enfado','text_Triste','text_MecCog','text_Insight','text_Causa','text_Discrep','text_Tentat','text_Certeza','text_Inhib','text_Incl','text_Excl','text_Percept','text_Ver','text_Oir','text_Sentir','text_Biolog','text_Cuerpo','text_Salud','text_Sexual','text_Ingerir','text_Relativ','text_Movim','text_Espacio','text_Tiempo','text_Trabajo','text_Logro','text_Placer','text_Hogar','text_Dinero','text_Relig','text_Muerte','text_Asentir','text_NoFluen','text_Relleno']

for key in LIWC_keys:
    df[key] = pd.to_numeric(df[key], errors='coerce')

# # Group by 'year', 'month', and 'day' and calculate the average of 'WC' and 'verbTU'
# result_df = df.groupby(['year', 'month', 'day'])[''].mean().reset_index()

# # Add a new column with the count of rows for each date
# result_df['count'] = df.groupby(['year', 'month', 'day'])['id'].transform('count')

result_df = df.groupby(['year', 'month', 'day'])[LIWC_keys].mean().reset_index()

# Calculate the count of rows for each date
count_df = df.groupby(['year', 'month', 'day'])['id'].count().reset_index()
count_df.rename(columns={'id': 'count'}, inplace=True)

# Merge the average data with the count data
result_df = pd.merge(result_df, count_df, on=['year', 'month', 'day'])

# Save the results to a new DataFrame
# result_df.to_csv('averages_by_date_with_count.csv', index=False)
print(result_df.head(50))
```
